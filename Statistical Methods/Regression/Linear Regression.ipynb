```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(-2, 2, 10000)

def f(x):
  return 1 - x**2+ x**3 # f(x) = 1-x^2

eps = np.random.normal(size=x.size) #random noise is normal. 
```

```python
y = f(x) + eps #eps = random noise
```

```python
plt.scatter(x, y, s=1)
```

```python
ones = np.ones_like(x)
X = np.column_stack([ones, x, x**2])
Theta = np.linalg.lstsq(X, y)[0] # leased square regression
```

```python
predicted = X @ Theta

plt.figure(figsize=[10, 5]) 
plt.scatter(x, f(x), s=3) # noise didnt efffect our preditions eventhough error shuold throw the prediction off
# plt.scatter(x, y, s=1)
plt.scatter(x, predicted, s=.1, color='r')
```
